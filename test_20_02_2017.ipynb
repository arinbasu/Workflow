## A Pleb's Guide to Use Nteract for Data analsis and writing articles

Nteract has the potential to be a replacement for [Jupyter notebooks](http://www.jupyter.org), although my first impressions were that this was far too lightweight compared with a jupyter notebook that sits on a browser. But it was a pleasant experience to work with a nteract notebook. It blends codes (bash shell, R, Jupyter, and Python -- should be sufficient with most data analysis work). It reads an ipynb notebook and can produce PDF, HTML outputs. I suppose it can use Pandoc to output into other formats. It may also work using a bash shell although one does not have to indicate with two % signs to start the process, just initiate a bash shell and that should do the trick.
